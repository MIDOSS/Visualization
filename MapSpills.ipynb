# Map the Oil Spills in Northern SoG Region

Spills between 49.5 and 50oN and West of -123.7

```python
import cmocean.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr

from salishsea_tools import viz_tools
```

```python
mesh = xr.open_dataset('/home/sallen/MEOPAR/grid/mesh_mask201702.nc')
tmask = 1- mesh.tmask[0, 0]
```

```python
data = pd.read_csv('/data/sallen/results/MIDOSS/Monte_Carlo/northern_strait.csv')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.pcolormesh(mesh.glamt[0], mesh.gphit[0], tmask, cmap=cm.topo)
ax.plot(data.spill_lon, data.spill_lat, 'o')
ax.set_xlim(-125.3, -123.7)
ax.set_ylim(49.4, 50.1)
```
